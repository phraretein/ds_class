```python
import bs4
import requests
```

```python
r = requests.get('https://rod.kaidee.com/product-351076017')
```

```python
html_page = bs4.BeautifulSoup(r.text, 'html.parser')
type(html_page)
```

```python
# BeautifulSoup.select return List[Element]
#html_page.select

# BeautifulSoup.select return Element object(First element)
#html_page.select_one
```

```python
# CSS Selector
selector = 'span.sc-1x47nlp-0.lbBTYY'
price_element = html_page.select_one(selector)
price_element_list = html_page.select(selector)
```

```python
# bs4.element.Tag.text
print(price_element)
price = price_element.text
price
```
