```python
import pandas as pd
from pythainlp.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.preprocessing import normalize
```

```python
df = pd.read_csv('/Users/phrare/ds_class/wk2_nlp/truevoice-intent/mari_train.csv')
df.head()
```

```python
word_tokenize('สวัสดีครับ')
```

```python
vectorizer = CountVectorizer(tokenizer=word_tokenize)
```

```python
X_train = vectorizer.fit_transform(df['texts'])
```

```python
words = vectorizer.get_feature_names()
```

```python
len(words)
#words
```

It is seen that there are total of 3577 words in our dictionary.

```python
y_train = df['destination'].astype('category').cat.codes
y_train
```

```python
clf = LinearSVC(random_state=0, tol=1e-5)
```

```python
clf.fit(X_train, y_train)  
```

```python
df_test = pd.read_csv('/Users/phrare/ds_class/wk2_nlp/truevoice-intent/mari_test.csv')
df_test.head()
```

```python
X_test = vectorizer.transform(df_test['texts'])
```

```python
y_test = df_test['destination'].astype('category').cat.codes
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

```python
X_train_norm = normalize(X_train, norm='l2')
X_test_norm = normalize(X_test, norm='l2')
```

```python
clf.fit(X_train_norm,y_train)
```

```python
clf.predict(X_test_norm)
```

```python
clf.score(X_test_norm, y_test)
```

It is seen that with normalization, the score increases. :)
